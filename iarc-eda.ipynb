# Set Environment

```python
import os
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from collections import Counter
```

# Reproducibility

```python
def set_seed(seed=19021):
    np.random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
set_seed()
```

# Metadata Frames from Excel

**Image Associated Data**

```python
image_df = pd.read_excel('/kaggle/input/bme-400-g03/IARCImageBankColpo/Cases - Images.xlsx')
image_df.head(20)
```

**Case Associated Data**

```python
case_df = pd.read_excel('/kaggle/input/bme-400-g03/IARCImageBankColpo/Cases Meta data.xlsx')
case_df.columns = case_df.iloc[0]
case_df = case_df[1:]
case_df.reset_index(drop=True, inplace=True)
case_df.head(20)
```

# Image Type Count

```python
types = Counter(image_df['Type'])
types
```

```python
values = [str(value) for value in types.keys()]
frequencies = list(types.values())

# Set font properties
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Computer Modern Roman']
plt.rcParams['font.size'] = 11

# Create bar plot
plt.bar(values, frequencies)
plt.xticks(rotation=90)
plt.xlabel('Types')
plt.ylabel('Frequencies')
plt.title('Histogram of Image Types')
plt.grid(axis='y', linestyle='--', alpha=0.75)

# Save the figure in high resolution
plt.savefig('image_types.png', dpi=600, bbox_inches='tight')

# Display the plot
plt.show()
```
